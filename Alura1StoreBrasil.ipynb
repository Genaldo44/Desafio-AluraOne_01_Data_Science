### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
loja['Faturamento'] = loja['Preço'] * loja['Quantidade de parcelas']
loja2['Faturamento'] = loja2['Preço'] * loja2['Quantidade de parcelas']
loja3['Faturamento'] = loja3['Preço'] * loja3['Quantidade de parcelas']
loja4['Faturamento'] = loja4['Preço'] * loja4['Quantidade de parcelas']

faturamento_total_loja1 = loja['Faturamento'].sum()
faturamento_total_loja2 = loja2['Faturamento'].sum()
faturamento_total_loja3 = loja3['Faturamento'].sum()
faturamento_total_loja4 = loja4['Faturamento'].sum()

print(f"Faturamento total Loja 1: R${faturamento_total_loja1:.2f}")
print(f"Faturamento total Loja 2: R${faturamento_total_loja2:.2f}")
print(f"Faturamento total Loja 3: R${faturamento_total_loja3:.2f}")
print(f"Faturamento total Loja 4: R${faturamento_total_loja4:.2f}")
```

# 2. Vendas por Categoria

```python
vendas_por_categoria_loja1 = loja.groupby('Categoria do Produto')['Faturamento'].sum().sort_values(ascending=False)
vendas_por_categoria_loja2 = loja2.groupby('Categoria do Produto')['Faturamento'].sum().sort_values(ascending=False)
vendas_por_categoria_loja3 = loja3.groupby('Categoria do Produto')['Faturamento'].sum().sort_values(ascending=False)
vendas_por_categoria_loja4 = loja4.groupby('Categoria do Produto')['Faturamento'].sum().sort_values(ascending=False)

vendas_por_categoria_total = pd.concat([vendas_por_categoria_loja1, vendas_por_categoria_loja2, vendas_por_categoria_loja3, vendas_por_categoria_loja4], axis=1)
vendas_por_categoria_total.columns = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

print("Vendas por Categoria (todas as lojas):")
display(vendas_por_categoria_total)
```

```python
todas_lojas = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

vendas_por_produto = todas_lojas['Produto'].value_counts()

produto_mais_vendido = vendas_por_produto.idxmax()
quantidade_mais_vendido = vendas_por_produto.max()

produto_menos_vendido = vendas_por_produto.idxmin()
quantidade_menos_vendido = vendas_por_produto.min()

print(f"Produto mais vendido: {produto_mais_vendido} ({quantidade_mais_vendido} unidades)")
print(f"Produto menos vendido: {produto_menos_vendido} ({quantidade_menos_vendido} unidades)")
```

# 3. Média de Avaliação das Lojas


```python
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f"Média de avaliação Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de avaliação Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de avaliação Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de avaliação Loja 4: {media_avaliacao_loja4:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
todas_lojas = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

vendas_por_produto = todas_lojas['Produto'].value_counts()

produto_mais_vendido = vendas_por_produto.idxmax()
quantidade_mais_vendido = vendas_por_produto.max()

produto_menos_vendido = vendas_por_produto.idxmin()
quantidade_menos_vendido = vendas_por_produto.min()

print(f"Produto mais vendido: {produto_mais_vendido} ({quantidade_mais_vendido} unidades)")
print(f"Produto menos vendido: {produto_menos_vendido} ({quantidade_menos_vendido} unidades)")
```

```python
lojas = {
    'Loja 1': loja,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4
}

for nome_loja, df_loja in lojas.items():
    vendas_por_produto_loja = df_loja['Produto'].value_counts()
    produto_mais_vendido_loja = vendas_por_produto_loja.idxmax()
    quantidade_mais_vendido_loja = vendas_por_produto_loja.max()
    produto_menos_vendido_loja = vendas_por_produto_loja.idxmin()
    quantidade_menos_vendido_loja = vendas_por_produto_loja.min()

    print(f"--- {nome_loja} ---")
    print(f"Produto mais vendido: {produto_mais_vendido_loja} ({quantidade_mais_vendido_loja} unidades)")
    print(f"Produto menos vendido: {produto_menos_vendido_loja} ({quantidade_menos_vendido_loja} unidades)")
    print("\n")
```
